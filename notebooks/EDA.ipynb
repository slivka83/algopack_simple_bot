# Импорт библиотек

```python
from moexalgo import Market, Ticker
import pandas as pd

import plotly.graph_objects as go
import matplotlib.pyplot as plt
import seaborn as sns
```

# Все акции

```python
# Все акции
stocks = Market('stocks')
```

## Торговая статистика

```python
# Данные по всем акциям за дату. Чтобы получить данные за период, нужно запустить цикл по датам
trd_df = stocks.tradestats(date='2023-10-10')
trd_df
```

```python
trd_df.sample(7).T
```

```python
trd_df['ticker'].nunique()
```

```python
temp_df = pd.pivot_table(trd_df, index=['ticker'] , columns=['tradetime'], values='trades', aggfunc='max', fill_value=0)
temp_df
```

```python
plt.figure(figsize=(17,15))
sns.heatmap(temp_df);
```

## Заказы

```python
# Данные по всем акциям за дату. Чтобы получить данные за период, нужно запустить цикл по датам
ord_df = stocks.orderstats(date='2023-10-10')
ord_df
```

## OBS

```python
# Данные по всем акциям за дату. Чтобы получить данные за период, нужно запустить цикл по датам
obs_df = stocks.obstats(date='2023-10-10').head()
obs_df
```

# Одна компания

```python
# Акции SBER
sber = Ticker('SBER')
```

## Свечи

```python
# Свечи по акциям SBER за период
sber_cndl_df = sber.candles(date='2023-11-25', till_date='2023-12-02', period='10m')
sber_cndl_df
```

```python
fig = go.Figure(data=[go.Candlestick(
    x=sber_cndl_df['begin'],
    open=sber_cndl_df['open'],
    high=sber_cndl_df['high'],
    low=sber_cndl_df['low'],
    close=sber_cndl_df['close']
)])
fig.show()
```

## Торговая статистика

По сути те же самые свечи но дополненные полезной инфрмацией.

```python
# Данные по акциям SBER за период
sber_trd_df = sber.tradestats(date='2023-11-25', till_date='2023-12-02')
sber_trd_df
```

```python
sber_trd_df.head(7).T
```

```python
fig = go.Figure(data=[go.Candlestick(
    x=sber_trd_df['systime'],
    open=sber_trd_df['pr_open'],
    high=sber_trd_df['pr_high'],
    low=sber_trd_df['pr_low'],
    close=sber_trd_df['pr_close']
)])
fig.show()
```
